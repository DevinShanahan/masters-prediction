```python
rmses = {}
rowans = {}
for m in [KNeighborsRegressor(37), Ridge(.1), RandomForestRegressor(min_samples_leaf=5, min_samples_split=2)]:
    rmse = []
    rowan = []
    print(m.__class__.__name__)
    for year in range(2004, 2022):
        y = stats[['ROWAN SCORE', 'PLAYER NAME']]  
        years = stats['YEAR']
        X = stats.drop(['ROWAN SCORE', 'PLAYER NAME', 'YEAR'], axis=1)
        trainX = X[years != year]
        testX = X[years == year]
        trainY = y[years != year]
        testY = y[years == year]
        poly = PolynomialFeatures()
        poly.fit(trainX)
        trainX = poly.transform(trainX)
        scaler = StandardScaler()
        scaler.fit(trainX)
        testX = poly.transform(testX)
        trainX = scaler.transform(trainX)
        testX = scaler.transform(testX)
        pca = PCA(4)
        pca.fit(trainX)
        trainX = pca.transform(trainX)
        testX = pca.transform(testX)
        names = testY['PLAYER NAME']
        trainY = trainY.drop('PLAYER NAME', axis=1)
        testY = testY.drop('PLAYER NAME', axis=1)
        m.fit(trainX, trainY.values.ravel())
        pred = m.predict(testX)
        ranks = stats['RANK_OFFICIAL_WORLD_GOLF_RANKING'][years == year]
        top15 = ranks <= 15
        top30 = (ranks > 15) & (ranks <= 30)
        rest = ranks > 30
        rowan1 = testY[top15][scipy.stats.rankdata(pred[top15]) <= 3].iloc[:3]
        assert len(rowan1) == 3
        rowan2 = testY[top30][scipy.stats.rankdata(pred[top30]) <= 3].iloc[:3]
        assert len(rowan2) == 3
        rowan3 = testY[rest][scipy.stats.rankdata(pred[rest]) <= 3].iloc[:3]
        assert len(rowan2) == 3
        print(year, (rowan1.sum() + rowan2.sum() + rowan3.sum())[0])
        rowan.append(rowan1.sum() + rowan2.sum() + rowan3.sum())
        rmse.append(mean_absolute_error(testY, pred) ** 0.5)
        testY['pred'] = pred
        testY['names'] = names
    
    rmses[m.__class__.__name__] = sum(rmse) / len(rmse)
    rowans[m.__class__.__name__] = sum(rowan) / len(rowan)
print(rmses)
print(rowans)
```
